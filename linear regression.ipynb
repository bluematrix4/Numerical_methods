```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(list(map(float, input("Enter values of x: ").split())))
y = np.array(list(map(float, input("Enter values of y: ").split())))
A=np.matrix([[len(x),np.sum(x)],[np.sum(x),np.sum(x**2)]])
X=np.matrix([[np.sum(y)],[np.sum(x*y)]])
b=np.linalg.solve(A,X)
```

```python
plt.scatter(x,y)
p=float(input("WHich x value do you want to predict y?  "))
f=lambda k: b[1,0]*k + b[0,0]
print("Value at",p,"is", f(p))
p=np.array([np.min(x),np.max(x)])
plt.plot(p,f(p),'-r')
print("y = ",b[0,0],"+",b[1,0],"x")
plt.show()
```
